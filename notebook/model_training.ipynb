```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("Algerian_forest_fires_dataset_cleaned.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(['day', 'month', 'year'],axis=1,inplace=True)
```

```python
df['Classes'].value_counts()
```

```python
df['Classes']=np.where(df['Classes'].str.contains("not fire"),0,1)
```

```python
df.head()
```

```python
df.tail()
```

```python
df['Classes'].value_counts()
```

```python
## divide the feature into independent and dependent
x=df.drop('FWI',axis=1)
y=df['FWI']
```

```python
x
```

```python
y
```

```python
## train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train.shape,x_test.shape
```

```python
## feature ssel based on corellation

x_train.corr()
```

```python
plt.figure(figsize=(12,10))
corr=x_train.corr()
sns.heatmap(corr,annot=True)
```

```python
## check for multi corlinearity
def correlation(dataset , thershold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>thershold:
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
## this will pick up the highly corellated (95%) and give me
# threshold is set above the thershold value it will return the value  
corr_features=correlation(x_train,0.85)
```

```python
## drop features when corellation is more corellated 
x_train.drop(corr_features,axis=1,inplace=True)
x_test.drop(corr_features,axis=1,inplace=True)
```

```python
##2 features has got droped on the basis of thershold
x_train.shape,x_test.shape
```

## feature scaling or standardation

```python
## feature scaling or standardation
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
x_train_scaled=scalar.fit_transform(x_train)
x_test_scaled=scalar.transform(x_test)
```

```python
x_train_scaled
```

## box plots to understand  effect of standard scalar

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,5))

# Left plot
plt.subplot(1,2,1)
sns.boxplot(data=x_train)
plt.title("x_train before scaling")

# Right plot
plt.subplot(1,2,2)
sns.boxplot(data=x_train_scaled)
plt.title("x_train after scaling")

plt.show()
```

## linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
linreg=LinearRegression()
linreg.fit(x_train_scaled,y_train)
y_pred=linreg.predict(x_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("mean absolute error",mae)
print("r2 score",score)
plt.scatter(y_test,y_pred)
```

## lasso regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
Lasso_reg=Lasso()
Lasso_reg.fit(x_train_scaled,y_train)
y_pred=Lasso_reg.predict(x_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("mean absolute error",mae)
print("r2 score",score)
plt.scatter(y_test,y_pred)
```

## cross validation lasso

## ridge regression model

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
Ridge_reg=Ridge()
Ridge_reg.fit(x_train_scaled,y_train)
y_pred=Ridge_reg.predict(x_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("mean absolute error",mae)
print("r2 score",score)
plt.scatter(y_test,y_pred)
```
